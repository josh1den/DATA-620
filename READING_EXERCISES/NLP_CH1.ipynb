### Counting Vocabulary

```python
# total characters aka tokens
len(text3)
```

```python
# obtain the vocabulary items aka unique tokens
set(text3)
```

```python
# sort the set
sorted(set(text3))
```

```python
# count lexcial richness
from __future__ import division 

len(text3) / len(set(text3))
```

```python
# how often does a word appear
text3.count("smote")
```

```python
# percentage of text
100 * text3.count("smote") / len(text3)
```

```python
# how many times does the word 'lol' appear in text 5
text5.count('lol')
```

```python
# how much is this as a percentage 
100 * text5.count('lol') / len(text5)
```

```python
# create lexical diversity function 
def lexical_diversity(text):
    return len(text) / len(set(text))

# create percentage function 
def percentage(count, total):
    return 100 * count / total
```

```python
lexical_diversity(text5)
```

```python
percentage(text5.count('lol'), len(text5))
```

### Finding Words Using Set Notation

```python
V = set(text1) 
long_words = [w for w in V if len(w) > 15]
sorted(long_words)[:5]
```

```python
# frequently occurring long words 
fdist5 = FreqDist(text5)
sorted([w for w in set(text5) if len(w) > 7 and fdist5[w] > 7])[:5]
```
