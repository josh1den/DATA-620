```python
# load module
import networkx as nx

# create Graph object 
G=nx.Graph()
```

We can pass a dict of lists containing edges and nodes:

```python
G = nx.from_dict_of_lists(
        {
            0: [1, 2, 3, 5],
            1: [0, 3, 4, 6],
            2: [0, 3, 5],
            3: [0, 1, 2, 4, 5, 6],
            4: [1, 3, 6],
            5: [0, 2, 3, 6, 7],
            6: [1, 3, 4, 5, 7],
            7: [5, 6, 8],
            8: [7, 9],
            9: [8],
        },
        create_using=None,
    )

G.name = "Krackhardt Kite Social Network"
nx.draw(G)
```

There's also a built-in function to do this. 

```python
import networkx.generators.small as nx1

g = nx1.krackhardt_kite_graph()

nx.draw(g)
```
