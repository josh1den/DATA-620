```python
# load modules
import networkx as nx

# initialize Graph object 
G=nx.Graph()
```

We can pass a dict of lists containing edges and nodes, where each key is a node, and each value are it's edges:

```python
G = nx.from_dict_of_lists(
        {
            0: [1, 2, 3, 5],
            1: [0, 3, 4, 6],
            2: [0, 3, 5],
            3: [0, 1, 2, 4, 5, 6],
            4: [1, 3, 6],
            5: [0, 2, 3, 6, 7],
            6: [1, 3, 4, 5, 7],
            7: [5, 6, 8],
            8: [7, 9],
            9: [8],
        },
        create_using=None,
    )

G.name = "Krackhardt Kite Social Network"
nx.draw(G)
```

There's also a built-in function to do this. 

```python
import networkx.generators.small as nx1

g = nx1.krackhardt_kite_graph()

nx.draw(g)
```

This can also be done by adding nodes and edges individually

```python
G1 = nx.Graph()
# add single node
G1.add_node(1)
```

```python
nx.draw(G1)
```

```python
G1.add_edge(1,2)
nx.draw(G1)
```

We can also pass node labels in as mappings and use optional arguments to make the graphic easier to read:

```python
mapping = {0: "Andre", 1: "Beverly", 2: "Diane", 3: "Carol", 4: "Ed", 5: "Fernando", 6: "Garth", 7: "Heather", 8: "Ike", 9: "Jane"}
G = nx.relabel_nodes(G, mapping)
nx.draw(G, with_labels = True, node_size=1500, node_color="#ADD8E6", edgecolors="#000000")
```
